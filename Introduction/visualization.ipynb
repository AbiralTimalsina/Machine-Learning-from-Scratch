## Why Choose Matplotlib and Seaborn
Both libraries are among the most popular choices for data visulization in Python. Here's why Matplotlib: Known for its flexibility, Matplotlib allows detailed customization of visual elements, making it perfect for tailored plots.

Seaborn: Built on top of Matplotlib, Seaborn simplifies visualization tasks by offering high level functions to create aesthetically pleasing statistical graphics.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Basics of matplotlib
# line plot

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='Sine wave')
plt.title('Line Plot Example')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Bar Chart

x = ['A', 'B', 'C',]
y = [10, 20, 30]
plt.bar(x, y, color=['blue', 'green', 'red'],  edgecolor='white')
# plt.style.use('dark_background')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()
```

```python
import seaborn as sns
```

```python
# histogram and KDE plot
data = np.random.randn(1000)
sns.histplot(data, kde=True, color='green')
plt.title('Histogram and KDE Plot Example')
plt.show()
```

```python
matrix = np.random.rand(5, 5)
sns.heatmap(matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
#pairplot example
sns.set_style("whitegrid")

iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
plt.show()
```

```python
df = pd.read_csv('Ice Cream Ratings.csv')
df
```

```python
df.index = df['Date']
df
```

```python
# scatter plot using texture rating, overall rating

sns.scatterplot(x='Texture Rating', y='Overall Rating', data=df)
plt.show()
```

```python
# hist plot

sns.histplot(x='Texture Rating', data=df)
plt.show()
```

```python
#box plot

sns.boxplot(x='Texture Rating', data=df)
plt.show()
```

```python
#pie chart
plt.pie(df['Texture Rating'], labels=df['Date'])
```
